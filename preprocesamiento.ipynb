# Datos de Lluvias 2019-2024

```python
pd.read_csv(f"data/lluvias_mes_a_mes/330113_{año}05_AguaCaida.csv", sep=";").head()
```

```python
#iterar por todos los años, y guardar en un array para armar un df, que tendra el año y la lluvia mensual promedio o algo así
lluvias = []
años = ["2019","2020","2021","2022","2023","2024"]
df = pd.DataFrame()
for año in años:
    try:
        mayo = pd.read_csv(f"data/lluvias_mes_a_mes/330113_{año}05_AguaCaida.csv", sep=";")
        # print("Mayo-",año,sep="")
    except:
        mayo = pd.DataFrame()
    try:
        junio = pd.read_csv(f"data/lluvias_mes_a_mes/330113_{año}06_AguaCaida.csv", sep=";")
        # print("Junio-",año,sep="")
    except:
        junio = pd.DataFrame()
    try:
        julio = pd.read_csv(f"data/lluvias_mes_a_mes/330113_{año}07_AguaCaida.csv", sep=";")
        # print("Julio-",año,sep="")
    except:
        julio = pd.DataFrame()
    try:
        ago = pd.read_csv(f"data/lluvias_mes_a_mes/330113_{año}08_AguaCaida.csv", sep=";")
        # print("Agosto-",año,sep="")
    except:
        ago = pd.DataFrame()

        


    df = pd.concat([mayo,junio,julio,ago])
    df = df[["momento","rrInst"]]
    df["fecha"] = df.momento.map(lambda x : x.split(" ")[0])
    suma_anual = df.groupby("fecha")["rrInst"].sum().sum() # Toma el la suma de todas las mediciones de una fecha para tener una por día, y luego suma la lluvia de los 4 meses
    lluvias.append(suma_anual) #Con media del año nos referimos a la media de la temporada de lluvia ya que si no la cantidad de datos a descargar y manejar sería mucha más y para epocas donde llueve poco
    print(año,": " ,suma_anual)
```

```python
df_f = pd.DataFrame(lluvias, columns=["sum_precip"])
df_f["año"] = años
df_f
```

```python
df_f.to_csv("data/suma_anual_precip.csv", index=False)
```
